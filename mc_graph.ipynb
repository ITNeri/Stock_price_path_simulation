```python
import math
import random
import matplotlib.pyplot as plt


def figure_path(s0, t, r, sig, m, n, z):
    # z refers to the number of lines in the graph, other variables are same as before
    delta_t = t / m  # length of time interval
    list_path = []

    for i in range(0, n):
        path = [s0]

        for j in range(0, m):
            path.append(
                path[-1] * math.exp((r - 0.5 * sig ** 2) * delta_t + (sig * math.sqrt(delta_t) * random.gauss(0, 1))))

        list_path.append(path)

    plt.figure(figsize=(12, 6))
    plt.grid(True)  # 显示网格线
    plt.title('stock price path simulation')
    plt.xlabel('time step')
    plt.ylabel('index level')
    plt.legend()

    for q in range(0, z):
        plt.plot(list_path[q])
```

```python
figure_path(100, 0.08, 0.03, 0.2, 100, 1000, 1000)
plt.show()
```
